## Gutenberg Corpus

```python
from nltk.corpus import gutenberg
```

```python
gutenberg.fileids()
```

```python
from tqdm import tqdm_notebook
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (15, 7)
import seaborn as sns
sns.set()
```

### Calculate Text Specifica

```python
t = tqdm_notebook(gutenberg.fileids())
result = []
for fid in t:
    t.set_description(fid)
    n_chars = len(gutenberg.raw(fid))
    n_words = len(gutenberg.words(fid))
    n_sents = len(gutenberg.sents(fid))
    n_vocab = len(set([w.lower() for w in gutenberg.words(fid)]))
    row = [int(n_chars/n_words), int(n_words/n_sents), int(n_words/n_vocab), fid]
    result.append(row)
df_result = pd.DataFrame.from_records(result, columns=["Chars/Words", "Words/Sents", "Words/Vocab", "Text"])
```

### Plot Text Specifica

```python
ax = df_result.iloc[:, [0, 1, 2]].plot.area(alpha=.25, stacked=False)
ax.set_xticks(df_result.index)
ax.set_xticklabels([t[:-4] for t in df_result["Text"]], rotation=45, ha="right");
```

## Lexical Ressources

### Words & Stopwords

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

```python
unusual_words(nltk.corpus.genesis.words("english-kjv.txt"))[:10]
```

```python
from nltk.corpus import stopwords
```

```python
def text_richness(text):
    stopw = stopwords.words("english")
    content = [w for w in text if w.lower() not in stopw]
    return len(content)/len(text)
```

```python
text_richness(nltk.corpus.reuters.words())
```

## A Pronouncing Dictionary

```python
entries = nltk.corpus.cmudict.entries()
```

```python
len(entries)
```

```python
for entry in entries[39943:39951]:
    print(entry)
```

```python
for word, proun in entries:
    if len(proun) == 3:
        p1, p2, p3 = proun
        if (p1 == 'P') and (p3 == 'T'):
            print(word, p2)
```

```python
syllable = ["N", "IH0", "K", "S"]
[word for word, pron in entries if pron[-4:] == syllable]
```

```python
[word for word, proun in entries if  (proun[-1] == "M") and (word[-1] == "n")]
```

```python
p3 = [(pron[0]+'-'+pron[2], word)
      for (word, pron) in entries
      if pron[0] == 'P' and len(pron) == 3]
```

```python
fdist_p3 = ConditionalFreqDist(p3)
```

```python
for template in fdist_p3.conditions():
    if (len(fdist_p3[template]) > 10):
        words = fdist_p3[template].keys()
        wordlist = " ".join(words)
        print(template, wordlist[:70])
```
